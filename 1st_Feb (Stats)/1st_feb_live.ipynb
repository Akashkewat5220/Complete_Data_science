```python
import cv2
```

```python
image = cv2.imread('WIN_20241129_16_48_18_Pro.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
u ,s ,vt = np.linalg.svd(image)
```

```python
len(u)
```

```python
compressed_image = np.dot(u[:, :70] ,np.dot(np.diag(s[:70]), vt[:70,:]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(compressed_image,cmap="gray")
```
